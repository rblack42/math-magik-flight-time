# Basic Propeller Theory

Obviously, the propulsion system in our indoor models plays a significant role in the flight times we can achieve. In this section, we will explore some of the basic theory into how a typical indoor model is powered during a flight. This theory has two primary components: the propeller itself, and the rubber "engine" used to drive that propeller.

## Rubber Motors

Let's start off by looking at our power source, the simple (well not so simple) wound-up rubber band!

In McLean's original article, the energy stored in a motor is a simple function of the model weight. Seasoned indoor flyers know this approximation is a bit weak. Winding techniques, rubber lubrication, and time held before launch can significantly affect the flight times achieved. Unfortunately, these variables are hard to incorporate into a simple flight time prediction scheme, so we will ignore them for the moment.

From experimental data on the rubber we use, Doug came up with this formula for the amount of energt **E** we can obtain from an indoor motor:

\begin{equation}
E = K_m w_m
\end{equation}

Where:
- k_m is a constant = 30000 inches
- w_m is the motor weight in grams








# Propellers

The propeller's job is pretty simple. Somehow, it has to *bite* into the air in front of the propeller and throw a chunk of air to the rear of that propeller at a higher velocity. The effect of that action will be to *thrust* the propeller, and anything connected to it, forward.

There are may ways we can cause this effect. Swimmers use their hands to generate *thrust* that moves them forward through their working fluid, water in this case. We often use our hands to do something similar, when we fan a fire to get the flames to go higher, for example. Obviously, flapping our hands is not going to turn us into birds, so we need something better.

Early researchers into flight settled on the idea of taking a wing-like surface called a *blade* attached to a shaft that is rotated by a power source. Usually propellers have two blades to make rotating them easier, but that is not a magic number. Single blade propellers are sometimes used, and more than two is not unheard of.

The shape of the blade is pretty important in generating he most thrust possible. If we view the blade motion as a model moves through the air, we see something like this:



```python
import matplotlib.pyplot as plt
import numpy as np

# Resizing Figure
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Creating Data
x = np.linspace(0 , 20 , 100)
y = np.sin(x)
z = np.cos(x)

# 3D Graphic
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_title('Prop Blade Tip Path')
ax.set_xlabel('path')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.tight_layout()
# 3D line plot
ax.plot3D(x, y, z)
```

This curve is called a *helix*, and it shows the blade tip advancing through the air as the propeller spins. The shape of the blade and its alignment on the propeller shaft determine how far the propeller advances on each revolution. 

If we assume that the propeller is moving through the air at a velocity given by **V**, then we can define something called the *advance ratio* of the propeller. Let **d** be the propeller diameter, and **n** be the number of revolutions per second we turn the propeller. Then the distance traveled on one revolution of the propeller is $V/n$. The *advance ratio* **J** ia the ratio of the distance traveled in one revolution to the diameter of the propeller.

\begin{equation}
J = \frac{V}{nd}
\end{equation}
